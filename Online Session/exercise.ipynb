```python
import pandas as pd

# Load your salesman sheet
df = pd.read_excel("Data Exercises 1.xlsx", sheet_name="Salesman")

# Standardize column names
df.columns = df.columns.str.strip().str.lower()

# Incentive definitions
VALUE_INC = 2000
EC_INC = 500
OUTLET_INC = 500
FIVESTAR_INC = 500

# Function to calculate incentive for each row
def calculate_incentive(row):
    total = 0
    breakdown = []

    if row['value'] == "Y":
        total += VALUE_INC
        breakdown.append("Value(2000)")
    
    if row['ec'] == "Y":
        total += EC_INC
        breakdown.append("EC(500)")
    
    if row['outlet'] == "Y":
        total += OUTLET_INC
        breakdown.append("Outlet(500)")
    
    if row['5star'] == "Y":
        total += FIVESTAR_INC
        breakdown.append("5Star(500)")

    return pd.Series([total, " + ".join(breakdown)])

# Apply function
df[['incentive_amount', 'breakdown']] = df.apply(calculate_incentive, axis=1)

print(df)
```

```python
import pandas as pd

# Load your Excel
df = pd.read_excel("Data Exercises 1.xlsx", sheet_name="PT#2(Basic)")

# Clean column names in case of spaces/newlines
df.columns = df.columns.str.strip()

# Function to calculate incentive
def calculate_incentive(row):
    incentive = 0
    
    # 2000 if value target achieved
    if str(row['Value_Target']).strip().lower() == "yes":
        incentive += 2000
    
    # 500 each for EC, Outlet Opening, 5-Star
    if str(row['EC_Target']).strip().lower() == "yes":
        incentive += 500

    if str(row['Outlet_Opening']).strip().lower() == "yes":
        incentive += 500

    if str(row['FiveStar_Target']).strip().lower() == "yes":
        incentive += 500

    return incentive

# Apply function to DataFrame
df['Incentive_Amount'] = df.apply(calculate_incentive, axis=1)

# Display result
print(df)

# Save to Excel
df.to_excel("salesman_incentives_output.xlsx", index=False)
print("\nFile saved as 'salesman_incentives_output.xlsx'")
```

```python
import pandas as pd
import unicodedata

def normalize(c):
    # fix unicode spaces
    c = ''.join(ch for ch in unicodedata.normalize("NFKC", c))
    c = c.strip().lower().replace(" ", "_")
    return c

df = pd.read_excel("Data Exercises 1.xlsx", sheet_name="PT#2(Basic)")

# normalize all columns
df.columns = [normalize(c) for c in df.columns]

print(df.columns.tolist())
```

```python
import pandas as pd

# Load the file
df = pd.read_excel("Data Exercises 1.xlsx", sheet_name="PT#2(Basic)")

# Clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# Rename properly if needed
df = df.rename(columns={
    "salesman_crsalesman_name": "salesman_name",
    "value_target": "value_target",
    "value_achievement": "value_achievement",
    "ec_achv": "ec_achv",
    "outlet_opening_tgt": "outlet_target",
    "outlet_opening_achv": "outlet_achv",
    "5_star_tgt": "star_target"
})

def calculate_incentive(row):
    incentive = 0
    breakdown = []

    # Value Target
    if row['value_achievement'] >= row['value_target']:
        incentive += 2000
        breakdown.append("Value(2000)")

    # EC achievement
    if row['ec_achv'] >= row['value_target']:   # using teacher logic mapping
        incentive += 500
        breakdown.append("EC(500)")

    # Outlet Achievement
    if row['outlet_achv'] >= row['outlet_target']:
        incentive += 500
        breakdown.append("Outlet(500)")

    # 5 Star Achievement (row has only target)
    if row['star_target'] >= 250:      # seen from your data pattern
        incentive += 500
        breakdown.append("5Star(500)")

    return pd.Series([incentive, " + ".join(breakdown)])

# Apply logic
df[['incentive_amount', 'breakdown']] = df.apply(calculate_incentive, axis=1)

print(df[['salesman_name', 'incentive_amount', 'breakdown']])
```

```python
import pandas as pd

# Load clean table by setting header row = 7
df = pd.read_excel("Data Exercises 1.xlsx", sheet_name="PT#2(Basic)", header=7)

# Remove fully empty rows
df = df.dropna(how="all").reset_index(drop=True)

# -----------------------------
# INCENTIVE CALCULATION LOGIC
# -----------------------------
def calc(row):
    incentive = 0
    breakdown = []

    # VALUE TARGET → if achievement >= target
    if row['Value achievement'] >= row['Value Target']:
        incentive += 2000
        breakdown.append("Value(2000)")

    # EC TARGET
    if row['EC Achv'] >= row['EC target']:
        incentive += 500
        breakdown.append("EC(500)")

    # OUTLET OPENING TARGET
    if row['Outlet opening Achv'] >= row['Outlet opening tgt']:
        incentive += 500
        breakdown.append("Outlet(500)")

    # 5 STAR TARGET
    if row['5 Star Achv'] >= row['5 Star Tgt']:
        incentive += 500
        breakdown.append("5Star(500)")

    return pd.Series([incentive, " + ".join(breakdown)])


# Apply logic
df[['Incentive', 'Breakdown']] = df.apply(calc, axis=1)

# Show final results
df[['SalesMan Code', 'SalesMan Name', 'Incentive', 'Breakdown']]
```
